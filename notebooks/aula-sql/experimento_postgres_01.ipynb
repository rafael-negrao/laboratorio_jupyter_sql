## Criando a tabela de produtos

```python
%%sql
CREATE TABLE
  IF NOT EXISTS produtos (
    id_produto SERIAL PRIMARY KEY,
    nome VARCHAR(120) NOT NULL,
    descricao TEXT,
    preco DECIMAL(10, 2) NOT NULL DEFAULT 0.00,
    ativo SMALLINT NOT NULL DEFAULT 1,
    criado_em TIMESTAMP DEFAULT CURRENT_TIMESTAMP
  )
```

## Fazendo um carga teste

```python
%%sql
INSERT INTO
  produtos (nome, descricao, preco, ativo)
VALUES
  (
    'PRODUTO TESTE',
    'SUPER DESCRICAO TESTE',
    10.99,
    1
  )
```

## Recuperando os dados via SQL

```python
%%sql
SELECT
  *
FROM
  produtos
```

```python
%%sql
SELECT
  ativo,
  COUNT(1) AS qtde
FROM
  produtos
GROUP BY
  ativo
```
